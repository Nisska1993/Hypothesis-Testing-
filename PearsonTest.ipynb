# Pearson Test
(pour les relations entre 2 variables continues)
Teste s'il existe une correlation entre les 2 variables continues

H0 : " Le montant du pourboire n'est pas corrélé au montant de l'addition"

# Libraries

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
from scipy.stats import pearsonr
```

# Data

```python
df= sns.load_dataset('tips')
print(df.head())
```

```python
sns.scatterplot(data=df,x="total_bill",y="tip")
```

```python
pearsonr(df['total_bill'],df['tip'])
```

```python
filtred_df =df.query("`sex`=='Female' & `day`=='Fri'")
```

```python
sns.scatterplot(data =filtred_df,x="total_bill",y="tip")
```

```python
pearsonr(filtred_df['total_bill'],filtred_df['tip'])
```

```python
print("H0 : I n' y a pas de correlation entre le total_bill et le tip")
print("H1 : Il y a une correlation entre le total_bill et le tip")

alpha = 0.02
p_value = pearsonr(filtred_df['total_bill'],filtred_df['tip']).pvalue
if p_value < alpha:
    print("On rejette H0")
else:
    print("On ne rejette pas H0")
```
